### Exercise 1

For $T = 1$, the individual's value function is:

\begin{equation}
\max\limits_{c_{1}\in[0, W_{1}]}u(c_{1})
\end{equation}

Plugging in the budget constraint $W_{2} = W_{1} - c_{1}$, we can rewrite the above equation in an equivalent way in terms of 
$W_{2}$:

\begin{equation}
\max\limits_{W_{2}\in[0, W_{1}]}u(W_{1} - W_{2})
\end{equation}

Since the individual lives only one period and $u(Â·)$ is increasing, he or she will choose $c_{1} = W_{1}$ and $W_{2} = 0$ to maximize the utility.

### Exercise 2

For $T = 2$, the individual's value function is:

\begin{equation}
\max\limits_{c_{1}\in[0, W_{1}],\; c_{2}\in[0, W_{2}]}u(c_{1}) + \beta u(c_{2})
\end{equation}

Using budget constraints to rewrite the above equation in terms of $W_{1}$, $W_{2}$ and $W_{3}$:

\begin{equation}
\max\limits_{W_{2}\in[0, W_{1}],\; W_{2}\in[0, W_{2}]}u(W_{1} - W_{2}) + \beta u(W_{2} - W_{3})
\end{equation}

Since the individual lives only two periods, by similar argument as before, he or she will choose $c_{2} = W_{2}$ and $W_{3} = 0$, $ {\forall}\;W_{1}, \; W_{2}$.

**So the condition that characterizes the optimal amount of cake to leave in period 2 is $W_{3} = 0$.**

Applying F.O.C. to the following equation w.r.t. $W_{2}$:

\begin{equation}
\max\limits_{W_{2}\in[0, W_{1}]}u(W_{1} - W_{2}) + \beta u(W_{2})
\end{equation}

we obtain **the condition that characterizes the optimal amount of cake to leave in period 1 is $-\frac{\mathrm{d} u}{\mathrm{d} (W_{2} - W_{1})}+\beta\frac{\partial u}{\partial W_{2}} = 0$.**

### Exercise 4

By the envelope theorem, $W_{T+1} = 0$. Applying the F.O.C., we obtain the condition that characterizes the optimal policy function in period $T - 1$ for $W_{T} = \psi_{T - 1}(W_{T - 1})$ as follows:

\begin{equation}
\frac{\partial u(W_{T - 1} - \psi_{T - 1}(W_{T - 1}))}{\partial W_{T - 1}} + \beta\frac{\partial u(\psi_{T - 1}(W_{T - 1})
)}{\partial W_{T - 1}}.
\end{equation}

Accordingly, the value function $V_{1}$ in terms of $\psi_{T - 1}(W_{T - 1})$ is

\begin{equation}
V_{T - 1}(W_{T - 1}) = u(W_{T - 1} -\psi_{T - 1}(W_{T - 1})) + \beta u(\psi_{T - 1}(W_{T - 1}))
\end{equation}

### Exercise 5

Given $u(c) = ln(c)$, we have

\begin{equation}
V_{T}(\overline W) = u(\overline W)  = ln(\overline W)
\end{equation}

\begin{equation}
\begin{split}
V_{T - 1}(\overline W) &= \max \limits_{W_{T}} u({\overline W} - W_{T}) + \beta u(W_{T}) \\
&= \max \limits_{W_{T}} ln({\overline W} - W_{T}) + \beta ln(W_{T})
\end{split}
\end{equation}

Taking F.O.C. to the second equation above w.r.t. $W_{T}$ we solve for $W_{T} = \frac{\beta}{\beta + 1}{\overline W}$.

Hence, $V_{T} = ln(\overline W) \neq V_{T-1} = ln\frac{1}{\beta + 1}{\overline W} + \beta ln\frac{\beta}{\beta + 1}{\overline W}$.

Moreover,

\begin{equation}
\psi_{T - 1}({\overline W}) = W_{T} = \frac{\beta}{\beta + 1}{\overline W}
\end{equation}
and 
\begin{equation}
\psi_{T}({\overline W}) = W_{T + 1} = 0.
\end{equation}

It is obvious from the above two equations that $\psi_{T - 1}({\overline W}) \neq \psi_{T}({\overline W})$.

### Exercise 6

**The finite horizon Bellman equation** is 

\begin{equation}
V_{T-2}(W_{T-2})=\max\limits_{W_{T-1}\in[0,\;W_{T-2}],\; W_{T}\in[0,\;W_{T-1}]}ln(W_{T-2}-W_{T-1})+\beta ln(W_{T-1}-W_{T})+\beta^2 ln(W_{T}-W_{T+1})
\end{equation}

By the envelope theorem, $W_{T+1} = 0$.

Taking F.O.C.s to the value function shown above w.r.t. $W_{T-1}$ and $W_{T}$ respectively,

\begin{equation}
-\frac{1}{W_{T-2}-W_{T-1}} + \beta \frac{1}{W_{T-1}-W_{T}} = 0
\end{equation}

\begin{equation}
-\beta \frac{1}{W_{T-1}-W_{T}} + \beta^2 \frac{1}{W_{T}} = 0
\end{equation}

we solve for $W_{T} = \frac{\beta}{1+\beta}W_{T-1} = \frac{\beta^2}{\beta^2 + \beta + 1}W_{T-1}$ and $W_{T-1} = \frac{\beta^2+\beta}{\beta^2+\beta+1}W_{T-2}$.

**Hence, the solution characterizes for the period $T-2$ policy function for how much cake to save for the next period is**

\begin{equation}
W_{T-1} = \psi_{T-2}(W_{T-2}) = \frac{\beta^2 + \beta}{\beta^2 + \beta + 1}W_{T-2} .
\end{equation}

Substituting $\{W_t\}^{T+1}_{t=T-2}$ back into the finite horizon Bellman equation, we get **the analytical solution for $V_{T-1}$** as follows:

\begin{equation}
V_{T-2}(W_{T-2}) = ln\frac{1}{\beta^2+\beta+1}W_{T-2} + \beta ln\frac{\beta}{\beta^2+\beta+1}W_{T-2} + \beta^2 ln\frac{\beta^2}{\beta^2+\beta+1}W_{T-2}
\end{equation}

### Exercise 7

By induction, **the analytical solution for $\psi_{T-S}(W_{T-S})$** is

\begin{equation}
\psi_{T-S}(W_{T-S}) = \frac{\sum^s_{i=1}\beta^i}{\sum^s_{i=0}\beta^i}W_{T-S}
\end{equation}

and **the analytical solution for $V_{T-S}(W_{T-S})$** is

\begin{equation}
V_{T-S}(W_{T-S}) = \sum_{i=0}^s \beta^i ln(\frac{\beta^i}{\sum_{j=0}^s \beta^j}W_{T-S}).
\end{equation}

Futhermore, when the horizon becomes infinity, we have

\begin{equation}
lim_{s\rightarrow\infty}\psi_{T-s}(W_{T-s}) = \beta W_{T-s} = \psi(W_{T-s})
\end{equation}

and

\begin{equation}
lim_{s\rightarrow\infty}V_{T-s}(W_{T-s}) = \frac{1}{1-\beta}ln((1-\beta)W_{T-s}) + \frac{\beta}{(1-\beta)^2}ln\beta = V(W_{T-s}).
\end{equation}

```python
from scipy.stats import norm
from scipy.integrate import quad
```

```python
def NC_discrete(mean, sigma, N, k):
    z = np.linspace(mean - k * sigma, mean + k * sigma, N)
    w = np.zeros(N)
    w[0] = norm.cdf((z[0] + z[1]) / 2, loc=mean, scale=sigma) 
    w[-1] = 1 - norm.cdf((z[-1] + z[-2]) / 2, loc=mean, scale=sigma)
    for i in range(1, N-1):
        z_min = (z[i-1] + z[i]) / 2
        z_max = (z[i] + z[i+1]) / 2
        norm_pdf = lambda x: norm.pdf(x, loc=mean, scale=sigma)
        w[i] = quad(norm_pdf, z_min, z_max)[0]
    return z, w
```

```python
sigma = np.sqrt(0.25)
mu = 4 * sigma
M = 7
z, w = NC_discrete(mu, sigma, M, k=3)
z, w
```

```python
v_t_plus_1_mat = np.zeros((N, M), dtype="float")
exp_v_t_plus_1_vec = np.dot(v_t_plus_1_mat, w.reshape(-1, 1)).reshape(1, 1, N)
exp_v_t_plus_1_arr = np.tile(exp_v_t_plus_1_vec, (N, M, 1))
exp_v_t_plus_1_arr[c_pos] = -1e10
s = 0

while 1:
    v_t_arr = u_arr + beta * exp_v_t_plus_1_arr
    v_t_opt_mat = v_t_arr.max(axis=2)
    dist = np.sum((v_t_opt_mat - v_t_plus_1_mat) ** 2)
    s += 1
    if dist < 1e-9:
        print("Interation:{}, distance:{}".format(s, dist))
        max_ind = np.argmax(v_t_arr, axis=2)
        p_t_opt_mat = np.zeros((N, M), dtype="float")
        for i in range(M):
            p_t_opt_mat[:, i] = w_vec[max_ind[:, i]]
        break
    elif s >= 1e5:
        print("Interation:{}, distance:{}".format(s, dist))
        print("The value function fails to converge after 10000 interations.")
        max_ind = np.argmax(v_t_arr, axis=2)
        p_t_opt_mat = np.zeros((N, M), dtype="float")
        for i in range(M):
            p_t_opt_mat[:, i] = w_vec[max_ind[:, i]]
        break
    else:
        print("Interation:{}, distance:{}".format(s, dist))
        v_t_plus_1_mat = v_t_opt_mat
        exp_v_t_plus_1_vec = np.dot(v_t_plus_1_mat, w.reshape(-1, 1)).reshape(1, 1, N)
        exp_v_t_plus_1_arr = np.tile(exp_v_t_plus_1_vec, (N, M, 1))
        exp_v_t_plus_1_arr[c_pos] = -1e10
```
